```python
import pandas as pd
```

```python
adult_data = pd.read_csv('Downloads/adult.data', header = None)
```

```python
adult_data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
                    'marital_status', 'occupation', 'relationship',
                    'race', 'sex', 'capital_gain', 'capital_loss',
                    'hours_per_week', 'native_country', 'income']
```

```python
for value in ['workclass', 'education',
          'marital_status', 'occupation',
          'relationship','race', 'sex',
          'native_country', 'income']:
    print (value,":", sum(adult_data[value] == '?'))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
La_En = LabelEncoder()
```

```python
workclass_num = La_En.fit_transform(adult_data.workclass)
education_num = La_En.fit_transform(adult_data.education)
marital_num   = La_En.fit_transform(adult_data.marital_status)
occupation_num = La_En.fit_transform(adult_data.occupation)
relationship_num = La_En.fit_transform(adult_data.relationship)
race_num = La_En.fit_transform(adult_data.race)
sex_num = La_En.fit_transform(adult_data.sex)
native_country_num = La_En.fit_transform(adult_data.native_country)
```

```python
adult_data['workclass_num_c'] = workclass_num
adult_data['education_num_c'] = education_num
adult_data['marital_num_c'] = marital_num
adult_data['occupation_num_c'] = occupation_num
adult_data['relationship_num_c'] = relationship_num
adult_data['race_num_c'] = race_num
adult_data['sex_num_c'] = sex_num
adult_data['native_country_num_c'] = native_country_num
```

```python
adult_data.head()
```

```python
dummy_columns = ['workclass', 'education',
          'marital_status', 'occupation',
          'relationship','race', 'sex',
          'native_country']

adult_data = adult_data.drop(dummy_columns, axis = 'columns')
```

```python
adult_data.head()
```

```python
adult_data = adult_data.reindex(['age', 'workclass_num_c', 'fnlwgt', 'education_num_c',
                                    'education_num','marital_num_c', 'occupation_num_c',
                                    'relationship_num_c', 'race_num_c', 'sex_num_c', 'capital_gain',
                                    'capital_loss', 'hours_per_week', 'native_country_num_c', 
                                    'income'], axis = 'columns')
```

```python
adult_data.head()
```

```python
adult_data.shape
```

```python
data_col = adult_data.columns
```

```python
feature = data_col.drop('income')
```

```python
feature
```

```python
scaled_feature = {}

for each in feature:
    mean, std = adult_data[each].mean(), adult_data[each].std()
    scaled_feature[each] = [mean, std]
    adult_data.loc[:, each] = (adult_data[each] - mean)/std
```

```python
adult_data.head()
```

```python
data_features = adult_data.iloc[:,:14]
data_target = adult_data.iloc[:,14]
```

```python
from sklearn.model_selection import train_test_split
```

```python
feature_train, feature_test, target_train, target_test = train_test_split(data_features, data_target, test_size = 0.3, random_state = 10)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(feature_train, target_train)
```

```python
target_predict = gnb.predict(feature_test)
```

```python
(target_predict == target_test).mean()
```
